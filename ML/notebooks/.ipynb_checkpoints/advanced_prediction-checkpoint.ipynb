# Water shed algo

```python
# Load the image and mask
# image = cv2.imread("D:\\Rooftop-Solar-Energy-Potential-Map\\ML\\data\\results\\source\\DL52_9309_8_3.jpg")
image = cv2.imread("D:\\Rooftop-Solar-Energy-Potential-Map\\ML\\data\\results\\mask\\DL54_55b2_3_6.jpg")
```

```python
img_grey = image[:,:,0]
plt.imshow(img_grey)
```

```python
ret1, thresh = cv2.threshold(img_grey, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
```

```python
plt.imshow(opening, cmap="gray")
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=10)
plt.imshow(sure_bg, cmap="gray")
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,0)
plt.imshow(dist_transform, cmap="gray")
```

```python
ret2, sure_fg = cv2.threshold(dist_transform, 0.025*dist_transform.max(),255,0)
plt.imshow(sure_fg, cmap="gray")
```

```python
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
plt.imshow(unknown, cmap="gray")
```

```python
ret3, markers = cv2.connectedComponents(sure_fg)
plt.imshow(markers)
```

```python
markers = markers+10
plt.imshow(markers)
```

```python
markers[unknown==255] = 0
plt.imshow(markers)
```

```python
markers = cv2.watershed(image, markers)
plt.imshow(markers , cmap ="gray")
```

```python
image[markers == -1] = [0,255,255] 
```

```python
img2 = color.label2rgb(markers, bg_label=0)
```

```python
plt.imshow(img2)
```

```python
props = measure.regionprops_table(markers, intensity_image=img_grey, 
                              properties=['label',
                                          'area', 'equivalent_diameter',
                                          'mean_intensity', 'solidity'])
```

```python
import pandas as pd
df = pd.DataFrame(props)
df = df[df.mean_intensity > 100]  #Remove background or other regions that may be counted as objects

print(df.head())
```
